```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

raw = yf.download ("SPY", start = "2025-01-01", end = "2026-01-01")
data = raw.droplevel("Ticker", axis = 1)
data['daily_return'] = data['Close'].pct_change()

print(data['daily_return'].describe())
```

```python
VaR_95 = data['daily_return'].quantile(0.05)
VaR_99 = data['daily_return'].quantile(0.01)

print(f"95% VaR: {VaR_95:.2%}")
print(f"99% VaR: {VaR_99:.2%}")
```

```python
fig, ax = plt.subplots(figsize=(12,5))
data['daily_return'].hist(bins=50, ax=ax, density=True,
                          alpha=0.7, label='Daily returns')

ax.axvline(VaR_95, color='orange', linewidth=2,
           linestyle='--', label=f'95% VaR: {VaR_95:.2%}')

ax.axvline(VaR_99, color='red', linewidth=2,
           linestyle='--', label=f'99% VaR: {VaR_99:.2%}')

ax.fill_betweenx([0, 50], VaR_99, VaR_95,
                 alpha=0.3, color='orange', label='5% tail')
ax.fill_betweenx([0,50], data['daily_return'].min(), VaR_99,
                 alpha=0.3, color='red', label='1% tail')

plt.title("SPY Daily Returns - VaR Thresholds")
plt.xlabel("Daily Return")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#data [Filter] - Python feature
CVaR_95 = data['daily_return'][data['daily_return'] <= VaR_95].mean()
CVaR_99 = data['daily_return'][data['daily_return'] <= VaR_99].mean()

print(f"95% CVaR: {CVaR_95:.2%}")
print(f"99% CVaR: {CVaR_99:.2%}")
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12,15))

# Chart 1 - Cumulative return
data['cumulative_return'] = (1 + data['daily_return']).cumprod()
(data['cumulative_return'] * 10000).plot(ax=axes[0])
axes[0].set_title("SPY - $10,000 invested Jan 2025")
axes[0].set_ylabel("Portfolio Value ($)")
axes[0].grid(True)

# Chart 2 - Rolling volatility
data['rolling_vol'] = data['daily_return'].rolling(21).std() * (252**0.5)
data['rolling_vol'].plot(ax=axes[1], color='orange')
axes[1].axhline(y=data['rolling_vol'].mean(), color='r',
                linestyle='--', label='Average vol')
axes[1].set_title("Rolling 21-day Volatility")
axes[1].set_ylabel("Annualised Vol")
axes[1].legend()
axes[1].grid(True)

# Chart 3 - Return distribution with VaR
data['daily_return'].hist(bins=50, ax=axes[2], density=True, alpha=0.7)
axes[2].axvline(VaR_95, color='orange', linewidth=2,
                linestyle='--', label=f":95% VaR: {VaR_95:.2%}")
axes[2].axvline(VaR_99, color='red', linewidth=2,
                linestyle='--', label=f":99% VaR: {VaR_99:.2%}")
axes[2].set_title("Daily Returns Distribution with VaR")
axes[2].set_ylabel("Density")
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.savefig('SPY_risk_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
```
